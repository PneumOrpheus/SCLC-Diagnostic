```python
# In the folder /home/data/Lung-PET-CT-Dx/ you will find the downloaded DICOM files. 
# I will now write code to convert these DICOM files to NIfTI format using the dcm2niix tool. Remove the dicom files after conversion to save space, and rename the nifti files to include the patient ID for easier identification.
import os
import subprocess
import shutil

# Path where data was downloaded
download_path = "/home/data/Lung-PET-CT-Dx/"

# Iterate over the metadata 'data' list from the previous cell
series = data[0]
print(series)
```

```python
i = 0
for series in data:
    i += 1
    print(f"Processing series {i} of {len(data)}")
    series_uid = series.get('SeriesInstanceUID')
    patient_id = series.get('PatientID')
    
    # tcia_utils typically creates a folder named after the SeriesInstanceUID
    dicom_dir = os.path.join(download_path, series_uid)
    
    if os.path.exists(dicom_dir):
        print(f"Processing Patient: {patient_id} (Series: {series_uid})...")
        
        # Construct the output filename: PatientID_SeriesUID (to ensure uniqueness)
        output_filename = f"{patient_id}_{series_uid}"
        if os.path.exists(os.path.join(download_path, output_filename + ".nii.gz")):
            print(f"  -> NIfTI file {output_filename}.nii.gz already exists. Skipping conversion.")
            continue
    
        
        # dcm2niix execution
        # -z y : Compress to .nii.gz
        # -f : Filename format
        # -o : Output directory
        # -b n : Do not create BIDS sidecar JSON
        # -i y : Ignore derived, localizer and 2D images
        cmd = [
            "dcm2niix",
            "-z", "y",
            "-f", output_filename,
            "-o", download_path,
            "-b", "n",
            "-i", "y",
            dicom_dir
        ]
        
        try:
            # Run conversion
            subprocess.run(cmd, check=True, stdout=subprocess.DEVNULL) # stdout=subprocess.DEVNULL silences the tool's verbose output
            
            # If successful, remove the original DICOM directory
            if os.path.exists(os.path.join(download_path, output_filename + ".nii.gz")): 
                shutil.rmtree(dicom_dir)
            print(f"  -> Converted to {output_filename}.nii.gz and removed source DICOMs.")
            
        except subprocess.CalledProcessError:
            print(f"  -> Error converting series {series_uid}")
        except Exception:
            print(f"  -> Unexpected error processing series {series_uid}")
    else:
        print(f"Directory for Series {series_uid} not found.")
```
